### Convert to Grayscale

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap = 'gray')
plt.axis("off") #remove axes ticks
plt.title('Grayscale Image')
```

### Canny Edge Detection

Canny Edge Detection is an algorithm used to detect edges in an image, and was developed by John F. Canny in 1986.

```python
rcParams['figure.figsize'] = 10, 12


edges = cv.Canny(img_corrected2, 
                  threshold1=100, 
                  threshold2=200)

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

### Histograms - Grayscale

```python
gray_img2 = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)

rcParams['figure.figsize'] = 20, 5
plt.subplot(121)
plt.axis("off") #remove axes ticks
plt.imshow(img_corrected2)

plt.subplot(122)
plt.hist(gray_img2.ravel(),256,[0,256])
plt.title('Histogram of bunny')
plt.show()
```

### RGB histogram

```python
rcParams['figure.figsize'] = 8, 4

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
